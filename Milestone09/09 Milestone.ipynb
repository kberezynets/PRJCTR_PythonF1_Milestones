
Task (do each point in separate cell in ipynb file ):
1) Read data from the file database.csv from milestone 5 into a data frame.
2) Add a Salary column, which is generated randomly from -10000 to 40000
3) Add a 'Correct salary' column with True value if Salary >= 0, else False value
4) If the Salary column in the data frame is < 0, change to 0
5) Save the new data frame to file database_with_salary.csv
6) For each "Department" sum the "Salary" and create a bar chart with all departments and the total salary

```python
# Task 1
# Read data from the file database.csv from milestone 5 into a data frame.

import pandas as pd

df = pd.read_csv("database.csv")  # reads file from filesystem

df
```

```python
# Task 2
# Add a Salary column, which is generated randomly from -10000 to 40000

import random

df["Salary"] = df.apply(lambda _: random.randint(-10_000, 40_000), axis=1)

df
```

```python
# Task 3
# Add a 'Correct salary' column with True value if Salary >= 0, else False value

df["Correct Salary"] = df["Salary"] >= 0

df
```

```python
# Task 4
# If the Salary column in the data frame is < 0, change to 0

# Replace negative values in the 'Salary' column with 0

df.loc[df['Salary'] < 0, 'Salary'] = 0

df
```

```python
# Task 5
# Save the new data frame to file database_with_salary.csv

df.to_csv("database_updated.csv")
```

```python
# Task 6
# For each "Department" sum the "Salary" and create a bar chart 
# with all departments and the total salary

import matplotlib.pyplot as plt

# Group by 'Department' and sum the 'Salary'
department_salaries = df.groupby('department')['Salary'].sum()

# Plot the total salaries by department
department_salaries.plot(kind='bar', color='skyblue')
plt.title("Total Salaries by Department")
plt.xlabel("Department")
plt.ylabel("Total Salary")
plt.xticks(rotation=45)
plt.show()
```
